```python
"""由于使用BeautifulSoup爬取网页有些内容爬取不到，这边参考了"杨松"同学使用xpath来爬取网页的方法"""
def get_subway_lines():
    html = etree.parse(url, etree.HTMLParser())
    html_data=html.xpath('//div[@class="line_place"]/div/table')

    subways = {}

    for data in html_data:
        line_name = data.xpath('./thead/tr[1]/td/text()')  # 获取地铁线名称
        line_name = re.sub(r'[首末班车时刻表 \r\n' ']+', '', line_name[0])  # re.sub()用于检索和替换==>处理字符
        stations = data.xpath('./tbody/tr/th[1]/text()')    # 获取对应站点名称
        stations = [s.strip() for s in stations]
        subways[line_name] = stations
    return subways

```

```python
## 获取站点相邻关系
def get_connction(subway):
    dicts = defaultdict(list)
  
    for line, stations in subway.items():
        for index, station in enumerate(stations):
            if index == 0:  # 第一站只与下一站相连
                dicts[station].append(stations[index + 1])
            elif index > 0 and index < len(stations) - 1:  # 中间的站和前后均相连
                dicts[station].append(stations[index-1])
                dicts[station].append(stations[index+1])
            else:   # 最后一站只与前一站相连
                dicts[station].append(stations[index-1])
    return dicts
```

```python
get_connction(get_subway_lines())
```

```python
import networkx as nx

"""显示中文地名"""
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


connection = get_connction(get_subway_lines())
subway_graph = nx.Graph(connection)
plt.figure(figsize=(16,16))  # 设置图片大小，figsize的参数分别是宽和高，单位英寸

"""nx.spring_layout建立布局，对图进行美化，用Fruchterman-Reingold算法排列节点，样子类似多中心放射状"""
nx.draw(subway_graph, pos=nx.spring_layout(subway_graph), with_labels=True, node_size=15)
```

```python
def is_goal(destination):
    def _wrap(current_path):
        return current_path[-1] == destination
    return _wrap

def search(graph, start, is_goal, search_strategy):
    pathes = [[start] ]
    seen = set()

    while pathes:
        path = pathes.pop(0)
        froniter = path[-1]

        if froniter in seen: continue

        successors = graph[froniter]

        for station in successors:
            if station in path: continue

            new_path = path + [station]
            pathes.append(new_path)

            if is_goal(new_path): return new_path

        seen.add(froniter)
        pathes = search_strategy(pathes)

def sort_path(cmp_func, beam=-1):
    def _sorted(pathes):
        return sorted(pathes, key=cmp_func)[:beam]
    return _sorted

def get_total_station(path):
    return len(path)

def get_as_much_path(path):
    return -1 * len(path)
```

```python
search(connection, start='古城', is_goal=is_goal('玉泉路'), search_strategy=lambda n:n)
```

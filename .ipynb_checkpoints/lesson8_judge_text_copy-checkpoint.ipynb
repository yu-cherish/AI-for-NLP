```python
x_title = content['title'].values
```

```python
y = content['is_xinhua'].values
```

```python
x_title[:10]
```

```python
y[:10]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=30000)   
x_title = vectorizer.fit_transform(x_title)  # 输出各个文本各个词的tfidf值===将原始文档集合转换为tf-idf特性矩阵。

# vectorizer = countVecotrizer()该类会将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j词在i类文本下的词频
# transformer = TfidfTransformer()该类会统计每个词语的tf-idf权值
# tfidf = transformer.fit_transform(vectorizer.fit_transform(x_title))
# 第一个fit_transform是计算tf-idf，第二个fit_transform是将文本转为词频矩阵
# 这三句话和上面的等价
```

```python
x_title
```

```python
x_title.shape
```

```python
indices = np.arange(len(y))
```

```python
indices
```

```python
np.random.shuffle(indices)
```

```python
indices
```

```python
split_test = 0.25
```

```python
train_indices = indices[int(len(indices)*split_test):]
test_indices = indices[:int(len(indices)*split_test)]
```

```python
X_t_train, X_t_test, y_train, t_test = (x_title[train_indices], x_title[test_indices], y[train_indices], y[test_indices])
```

```python
X_t_train
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()  #创建分类器，把GaussianNB赋值给clf（分类器）
clf.fit(X_t_train.toarray(), y_train) #fit替代了train，这里是我们实际提供训练数据的地方
#它会学习各种模式，然后就形成了我们刚刚创建的分类器（clf）
#我们在分类器上调用fit函数，接下来将两个参数传递给fit函数，一个是特征x一个是标签y
#在监督分类中这个过程都是如此，先调用fit函数，然后依次获得特征和标签
```

```python
clf.predict(X_t_test.toarray())  # 利用测试集进行预测
```

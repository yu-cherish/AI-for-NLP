```python
import requests
import re
import urllib.request


b_url='https://baike.baidu.com/'




def get_url(url,pattern):


    response =requests.get(url,headers='')
    print(response)
    print(url)

#     response.encoding='utf-8'
    print(response.text)
    pattern_get=re.compile(pattern)
    print(pattern_get)
    result=pattern_get.findall(response.text)
    print(result)

    return result

def get_links():
    url='item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485'
    links=get_url(b_url+url,'< a target=_blank href = "/item/.+?" >北京地铁.+?线</ a>') #“.”表示任意字符。“+”表示前面表达式一次乃至多次。“?”表示匹配模式是非贪婪的

    links_list={}
    for name,herf in links:
        links_list[name]=b_url+herf
    return links_list
```
